Pre-processing pickle files to extract caption

```python
main_path = "ikea-master"
```

```python
# Define paths
import pickle
text_path = os.path.join(main_path,'ikea-master/text_data')
save_path = os.path.join(main_path,'ikea-master/pickle_processed')

pickle_files = os.listdir(text_path)
```

```python
desired_order = [
    'products_dict.p',
    'img_to_desc.p',
    'categories_images_dict.p',
    'item_to_room.p',
    'room_to_items.p',
    'categories_dict.p'
]
pickle_files_sorted = sorted(pickle_files, key=lambda x: desired_order.index(x))
pickle_files_sorted
```

```python
for pickle_file in [pickle_files_sorted[0]]:
    print('extracting {} pickle file'.format(pickle_file))
    with open(os.path.join(text_path,pickle_file), 'rb') as file:
        deserialized_data = pickle.load(file)

annotations = []
for item_id, details in deserialized_data.items():
    annotations.append({
        "file_name" : item_id+'.jpg',
        # "desc" : ". ".join([details['desc']])
        "desc" : ". ".join([details['type'], details['color'],details['size'],details['desc']])
    })

with open(os.path.join(save_path,pickle_file.replace('.p','.json')), 'w') as annotations_file:
    json.dump(annotations, annotations_file, indent=4)

print(f"Annotations file created at: {save_path}")
        
```

```python
for pickle_file in pickle_files_sorted[:2]:
    print('extracting {} pickle file'.format(pickle_file))
    with open(os.path.join(text_path,pickle_file), 'rb') as file:
        deserialized_data = pickle.load(file)

annotations  = []                                            
for item_id, details  in deserialized_data.items():
    file_name = item_id.split('/')[-1]
    annotations.append({
        "file_name" : file_name,
        "desc" : details['desc']
    })
    
with open(os.path.join(save_path,pickle_file.replace('.p','.json')), 'w') as annotations_file:
    json.dump(annotations, annotations_file, indent=4)

print(f"Annotations file created at: {save_path}")
```

```python
# pip install accelerate
import requests
from PIL import Image
from transformers import Blip2Processor, Blip2ForConditionalGeneration

processor = Blip2Processor.from_pretrained("Salesforce/blip2-opt-2.7b")
model = Blip2ForConditionalGeneration.from_pretrained("Salesforce/blip2-opt-2.7b")

img_url = 'https://storage.googleapis.com/sfr-vision-language-research/BLIP/demo.jpg' 
raw_image = Image.open(requests.get(img_url, stream=True).raw).convert('RGB')

question = "how many dogs are in the picture?"
inputs = processor(raw_image, question, return_tensors="pt").to("cuda")

out = model.generate(**inputs)
print(processor.decode(out[0], skip_special_tokens=True))
```
